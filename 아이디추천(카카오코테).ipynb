```python
# 알파벳 소문자, 숫자, 빼기(-), 밑줄(_), 마침표(.)를 제외한 모든 문자를 제거합니다.
def isValid(c):
  if c.isalnum(): # c in 'abcdefghijklmnopqrstuvwunsyz'
    return True
  if c == '-' or c == '_' or c == '.':
    return True
  return False

def solution(new_id):
  answer = ''
  lastDot = False
  for ch in new_id: # 알파벳 소문자, 숫자, 빼기(-), 밑줄(_), 마침표(.)를 제외한 모든 문자를 제거합니다.
    if isValid(ch) == False:
      continue

    if ch == '.': # 마침표(.)가 2번 이상 연속된 부분을 하나의 마침표(.)로 치환합니다.
      if len(answer) == 0 or lastDot: # 마침표(.)가 처음에 위치하면 제거합니다.
        continue
      lastDot = True
    else:
      lastDot = False

    ch = ch.lower()
    answer += ch

  if len(answer) >= 16: # 길이가 16자 이상이면, 15개의 문자를 제외한 나머지를 제거합니다.
    answer = answer[:15]

  if answer.endswith('.'): # 마침표(.)가 끝에 위치한다면 제거합니다.
    answer = answer[:-1]

  if len(answer) == 0: # 빈 문자열이면 'a' 대입
    answer += 'a'

  if len(answer) <= 2: # 길이가 2자 이하라면, 마지막 문자를 길이가 3이 될 때까지 반복해서 끝에 붙인다.
    ch = answer[-1:]
    while len(answer) < 3:
      answer += ch

  return answer

no = [
    "...!@BaT#*..y.abcdefghijklm",
    "z-+.^.",
    "=.=",
    "123_.def",
    "abcdefghijklmn.p"
]

for i in no:
  print(solution(i))
```

# 알파벳 소문자, 숫자, 빼기(-), 밑줄(_), 마침표(.)를 제외한 모든 문자를  제거합니다.
# 마침표(.)가 2번 이상 연속된 부분을 하나의 마침표(.)로 치환합니다.
# 마침표(.)가 처음에 위치하면 제거합니다.
# 마침표(.)가 끝에 위치한다면 제거합니다.
# 길이가 16자 이상이면, 15개의 문자를 제외한 나머지를 제거합니다.
# 빈 문자열이면 'a' 대입
# 길이가 2자 이하라면, 마지막 문자를 길이가 3이 될 때까지 반복해서 끝에 붙인다.

```python
def solution(new_id):
  answer = ''
  
  contiDot = False
  for ch in new_id:
    if ch.isalnum() == False:
      if ch != '_':
        if ch != '.':
          if ch != '-':
            continue
    if ch == '.':
      if len(answer) == 0 or contiDot:
        contiDot = True
        continue
      contiDot = True
    else:
      contiDot = False

    ch = ch.lower()
    answer += ch

  if len(answer) > 15:
    answer = answer[:15]
  if answer[-1:] == '.':
    answer = answer[:-2]
  if answer == '':
    answer += 'a'
  if len(answer) <= 2:
    ch = answer[-1:]
    while len(answer) <= 2:
      answer += ch

  return answer

no = [
    "...!@BaT#*..y.abcdefghijklm",
    "z-+.^.",
    "=.=",
    "123_.def",
    "abcdefghijklmn.p"
]

for i in no:
  print(solution(i))
```
